```python
import pandas as pd
import matplotlib.pyplot as plt

yolo_cls_metrics = pd.read_csv('results_YOLO_3cls.csv')
yolo_metrics = pd.read_csv('results_YOLO.csv')

yolo_metrics.columns = yolo_metrics.columns.str.strip()
yolo_cls_metrics.columns = yolo_cls_metrics.columns.str.strip()

final_iou_yolo_map50_b = yolo_metrics['metrics/mAP50(B)'].iloc[-1]
final_iou_yolo_map50_95_b = yolo_metrics['metrics/mAP50-95(B)'].iloc[-1]
final_iou_yolo_map50_m = yolo_metrics['metrics/mAP50(M)'].iloc[-1]
final_iou_yolo_map50_95_m = yolo_metrics['metrics/mAP50-95(M)'].iloc[-1]

final_iou_yolo_cls_map50_b = yolo_cls_metrics['metrics/mAP50(B)'].iloc[-1]
final_iou_yolo_cls_map50_95_b = yolo_cls_metrics['metrics/mAP50-95(B)'].iloc[-1]
final_iou_yolo_cls_map50_m = yolo_cls_metrics['metrics/mAP50(M)'].iloc[-1]
final_iou_yolo_cls_map50_95_m = yolo_cls_metrics['metrics/mAP50-95(M)'].iloc[-1]

final_precision_yolo_b = yolo_metrics['metrics/precision(B)'].iloc[-1]
final_precision_yolo_m = yolo_metrics['metrics/precision(M)'].iloc[-1]

final_precision_yolo_cls_b = yolo_cls_metrics['metrics/precision(B)'].iloc[-1]
final_precision_yolo_cls_m = yolo_cls_metrics['metrics/precision(M)'].iloc[-1]

final_recall_yolo_b = yolo_metrics['metrics/recall(B)'].iloc[-1]
final_recall_yolo_m = yolo_metrics['metrics/recall(M)'].iloc[-1]

final_recall_yolo_cls_b = yolo_cls_metrics['metrics/recall(B)'].iloc[-1]
final_recall_yolo_cls_m = yolo_cls_metrics['metrics/recall(M)'].iloc[-1]

plt.figure(figsize=(12, 6))
plt.plot(yolo_metrics['epoch'], yolo_metrics['metrics/mAP50(B)'], label=f'YOLO mAP50 (B) - Final: {final_iou_yolo_map50_b:.2f}', color='blue')
plt.plot(yolo_metrics['epoch'], yolo_metrics['metrics/mAP50-95(B)'], label=f'YOLO mAP50-95 (B) - Final: {final_iou_yolo_map50_95_b:.2f}', color='cyan')
plt.plot(yolo_metrics['epoch'], yolo_metrics['metrics/mAP50(M)'], label=f'YOLO mAP50 (M) - Final: {final_iou_yolo_map50_m:.2f}', color='skyblue')
plt.plot(yolo_metrics['epoch'], yolo_metrics['metrics/mAP50-95(M)'], label=f'YOLO mAP50-95 (M) - Final: {final_iou_yolo_map50_95_m:.2f}', color='navy')

plt.plot(yolo_cls_metrics['epoch'], yolo_cls_metrics['metrics/mAP50(B)'], label=f'YOLO 3 clases mAP50 (B) - Final: {final_iou_yolo_cls_map50_b:.2f}', color='green')
plt.plot(yolo_cls_metrics['epoch'], yolo_cls_metrics['metrics/mAP50-95(B)'], label=f'YOLO 3 clases mAP50-95 (B) - Final: {final_iou_yolo_cls_map50_95_b:.2f}', color='lightgreen')
plt.plot(yolo_cls_metrics['epoch'], yolo_cls_metrics['metrics/mAP50(M)'], label=f'YOLO 3 clases mAP50 (M) - Final: {final_iou_yolo_cls_map50_m:.2f}', color='darkgreen')
plt.plot(yolo_cls_metrics['epoch'], yolo_cls_metrics['metrics/mAP50-95(M)'], label=f'YOLO 3 clases mAP50-95 (M) - Final: {final_iou_yolo_cls_map50_95_m:.2f}', color='limegreen')
plt.xlabel('Epoch')
plt.ylabel('mAP ')
plt.title('Comparison of Validation mAP')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(yolo_metrics['epoch'], yolo_metrics['metrics/precision(B)'], label=f'YOLO Precision (B) - Final: {final_precision_yolo_b:.2f}', color='purple')
plt.plot(yolo_metrics['epoch'], yolo_metrics['metrics/precision(M)'], label=f'YOLO Precision (M) - Final: {final_precision_yolo_m:.2f}', color='magenta')

plt.plot(yolo_cls_metrics['epoch'], yolo_cls_metrics['metrics/precision(B)'], label=f'YOLO 3 clases Precision (B) - Final: {final_precision_yolo_cls_b:.2f}', color='red')
plt.plot(yolo_cls_metrics['epoch'], yolo_cls_metrics['metrics/precision(M)'], label=f'YOLO 3 clases Precision (M) - Final: {final_precision_yolo_cls_m:.2f}', color='orange')
plt.xlabel('Epoch')
plt.ylabel('Precision')
plt.title('Training Precision Comparison')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(yolo_metrics['epoch'], yolo_metrics['metrics/recall(B)'], label=f'YOLO Recall (B) - Final: {final_recall_yolo_b:.2f}', color='blue')
plt.plot(yolo_metrics['epoch'], yolo_metrics['metrics/recall(M)'], label=f'YOLO Recall (M) - Final: {final_recall_yolo_m:.2f}', color='cyan') 

plt.plot(yolo_cls_metrics['epoch'], yolo_cls_metrics['metrics/recall(B)'], label=f'YOLO 3 clases Recall (B) - Final: {final_recall_yolo_cls_b:.2f}', color='limegreen')
plt.plot(yolo_cls_metrics['epoch'], yolo_cls_metrics['metrics/recall(M)'], label=f'YOLO 3 clases Recall (M) - Final: {final_recall_yolo_cls_m:.2f}', color='darkgreen')
plt.xlabel('Epoch')
plt.ylabel('Recall')
plt.title('Training Recall Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
