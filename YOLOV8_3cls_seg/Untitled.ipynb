```python
import cv2
from ultralytics import YOLO
import numpy as np

# Cargar el modelo entrenado
model = YOLO("/home/voicelab/Desktop/segmentation_glottis/YOLOV8_3cls_seg/runs/segment/train/weights/best.pt")

# Ruta al video de prueba
video_path = "/home/voicelab/Desktop/segmentation_glottis/videos_VPLab/FN001.avi"
cap = cv2.VideoCapture(video_path)

# Obtener información del video
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# Crear un objeto VideoWriter para guardar el video de salida
output_path = "/home/voicelab/Desktop/segmentation_glottis/YOLOV8_3cls_seg/videos_segmentados/FN001_version2.avi"
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*"MJPG"), fps, (w, h))

# Colores para las clases
COLORS = {
    1: (255, 0, 0),       
    2: (0, 255, 0),       
    3: (0, 0, 255)        
}

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Video frame is empty or video processing has been successfully completed.")
        break

    # Ejecutar inferencia en el fotograma
    results = model(frame)
    
    if results[0].masks is not None:
        masks = results[0].masks
        for mask, cls in zip(masks.xy, results[0].boxes.cls):
            cls = int(cls)   # Asegurarse de que la clase esté bien indexada
            color = COLORS.get(cls, (0, 255, 255))  # Amarillo para clases no definidas
            mask = mask.astype(np.int32)
            cv2.polylines(frame, [mask], isClosed=True, color=color, thickness=2)

    out.write(frame)
    cv2.imshow("YOLOv8 Inference", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Liberar los recursos
cap.release()
out.release()
cv2.destroyAllWindows()
```
