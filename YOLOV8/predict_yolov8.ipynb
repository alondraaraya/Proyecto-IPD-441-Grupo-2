```python
import cv2
from ultralytics import YOLO
import numpy as np

# Se carga el modelo YOLO con el peso entrenado
model = YOLO("/home/voicelab/Desktop/segmentation_glottis/YOLOV8/runs/segment/train2/weights/best.pt")

# Se define ruta del video a segmentar
video_path = "/home/voicelab/Desktop/segmentation_glottis/videos_VPLab/FN009_ROI.mp4"
cap = cv2.VideoCapture(video_path)

# Se obtienen las propiedades del video
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# Se define ruta del video a segmentado
output_path = "/home/voicelab/Desktop/segmentation_glottis/YOLOV8/videos_segmentados_yolo/FN009_ROI_seg_YOLO_1cls.avi"
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*"MJPG"), fps, (w, h))

# Se define el color dverde para los contornos
green_color = (0, 255, 0)

# Se procesa cada frame del video
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:# Si el frame esta vacio el video termino 
        print("Video frame is empty or video processing has been successfully completed.")
        break
        
    # Se aplica el modelo yolo al frame par arealizar la prediccion
    results = model(frame)
    
    if results[0].masks is not None:# Si la prediccion arroja una mascara
        masks = results[0].masks.xy # Coordenadas de la mascara
        for mask in masks:
            mask = mask.astype(np.int32)  
            # Dibujar los contornos
            cv2.polylines(frame, [mask], isClosed=True, color=green_color, thickness=2)

    out.write(frame)
    cv2.imshow("YOLOv8 Inference", frame) # Mostrar el frame del video

    if cv2.waitKey(1) & 0xFF == ord("q"): # salir del loop si se apreta q
        break
# Liberar recursos
cap.release()
out.release()
cv2.destroyAllWindows()
```
